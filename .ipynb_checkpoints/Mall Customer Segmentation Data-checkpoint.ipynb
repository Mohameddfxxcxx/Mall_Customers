```python
import numpy as np  # For numerical computations
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns  # For statistical data visualization
```

```python
# Load the dataset "Mall_Customers.csv" into a DataFrame named 'df'
df = pd.read_csv("Mall_Customers.csv")
```

```python
# Display concise summary of the DataFrame 'df'
df.info()
```

```python
# Display the column names of the DataFrame 'df'
df.columns
```

```python
# Display the first few rows of the DataFrame 'df'
df.head()
```

```python
# Drop the column 'CustomerID' from the DataFrame 'df'
df.drop(["CustomerID"], axis=1, inplace=True)
```

```python
# Create a new figure with a specific size
plt.figure(figsize=(10, 6))

# Set the title of the plot
plt.title("Ages Frequency")

# Set the seaborn style to "dark"
sns.axes_style("dark")

# Create a violin plot of the 'Age' column
sns.violinplot(y=df["Age"])

# Display the plot
plt.show()
```

```python
# Create a new figure with a specific size
plt.figure(figsize=(15, 6))

# Create the first subplot
plt.subplot(1, 2, 1)

# Create a boxplot of the 'Spending Score (1-100)' column with red color
sns.boxplot(y=df["Spending Score (1-100)"], color="red")

# Create the second subplot
plt.subplot(1, 2, 2)

# Create a boxplot of the 'Annual Income (k$)' column
sns.boxplot(y=df["Annual Income (k$)"])

# Display the plot
plt.show()
```

```python
# Count the number of occurrences of each gender in the 'Gender' column
genders = df.Gender.value_counts()

# Set the seaborn style to "darkgrid"
sns.set_style("darkgrid")

# Create a new figure with a specific size
plt.figure(figsize=(10, 4))

# Create a bar plot showing the counts of each gender
sns.barplot(x=genders.index, y=genders.values)

# Display the plot
plt.show()
```

```python
# Selecting age groups using boolean indexing
age18_25 = df.Age[(df.Age <= 25) & (df.Age >= 18)]
age26_35 = df.Age[(df.Age <= 35) & (df.Age >= 26)]
age36_45 = df.Age[(df.Age <= 45) & (df.Age >= 36)]
age46_55 = df.Age[(df.Age <= 55) & (df.Age >= 46)]
age55above = df.Age[df.Age >= 56]

# Define age group labels and counts
x = ["18-25", "26-35", "36-45", "46-55", "55+"]
y = [len(age18_25.values), len(age26_35.values), len(age36_45.values), len(age46_55.values), len(age55above.values)]

# Create a new figure for the plot with a specific size
plt.figure(figsize=(15, 6))

# Create a bar plot showing the number of customers in each age group
sns.barplot(x=x, y=y, palette="rocket")

# Set the title, x-axis label, and y-axis label
plt.title("Number of Customers by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")

# Display the plot
plt.show()
```

```python
# Selecting spending score groups using boolean indexing
ss1_20 = df["Spending Score (1-100)"][(df["Spending Score (1-100)"] >= 1) & (df["Spending Score (1-100)"] <= 20)]
ss21_40 = df["Spending Score (1-100)"][(df["Spending Score (1-100)"] >= 21) & (df["Spending Score (1-100)"] <= 40)]
ss41_60 = df["Spending Score (1-100)"][(df["Spending Score (1-100)"] >= 41) & (df["Spending Score (1-100)"] <= 60)]
ss61_80 = df["Spending Score (1-100)"][(df["Spending Score (1-100)"] >= 61) & (df["Spending Score (1-100)"] <= 80)]
ss81_100 = df["Spending Score (1-100)"][(df["Spending Score (1-100)"] >= 81) & (df["Spending Score (1-100)"] <= 100)]

# Define spending score group labels and counts
ssx = ["1-20", "21-40", "41-60", "61-80", "81-100"]
ssy = [len(ss1_20.values), len(ss21_40.values), len(ss41_60.values), len(ss61_80.values), len(ss81_100.values)]

# Create a new figure for the plot with a specific size
plt.figure(figsize=(15, 6))

# Create a bar plot showing the number of customers in each spending score group
sns.barplot(x=ssx, y=ssy, palette="nipy_spectral_r")

# Set the title, x-axis label, and y-axis label
plt.title("Spending Scores")
plt.xlabel("Score")
plt.ylabel("Number of Customers Having the Score")

# Display the plot
plt.show()
```

```python
# Selecting annual income groups using boolean indexing
ai0_30 = df["Annual Income (k$)"][(df["Annual Income (k$)"] >= 0) & (df["Annual Income (k$)"] <= 30)]
ai31_60 = df["Annual Income (k$)"][(df["Annual Income (k$)"] >= 31) & (df["Annual Income (k$)"] <= 60)]
ai61_90 = df["Annual Income (k$)"][(df["Annual Income (k$)"] >= 61) & (df["Annual Income (k$)"] <= 90)]
ai91_120 = df["Annual Income (k$)"][(df["Annual Income (k$)"] >= 91) & (df["Annual Income (k$)"] <= 120)]
ai121_150 = df["Annual Income (k$)"][(df["Annual Income (k$)"] >= 121) & (df["Annual Income (k$)"] <= 150)]

# Define annual income group labels and counts
aix = ["$ 0 - 30,000", "$ 30,001 - 60,000", "$ 60,001 - 90,000", "$ 90,001 - 120,000", "$ 120,001 - 150,000"]
aiy = [len(ai0_30.values), len(ai31_60.values), len(ai61_90.values), len(ai91_120.values), len(ai121_150.values)]

# Create a new figure for the plot with a specific size
plt.figure(figsize=(15, 6))

# Create a bar plot showing the number of customers in each annual income group
sns.barplot(x=aix, y=aiy, palette="Set2")

# Set the title, x-axis label, and y-axis label
plt.title("Annual Incomes")
plt.xlabel("Income")
plt.ylabel("Number of Customers")

# Display the plot
plt.show()
```

```python
# Importing necessary module for 3D plotting
from mpl_toolkits.mplot3d import Axes3D

# Setting seaborn style to "white"
sns.set_style("white")

# Creating a new figure for the plot with a specific size
fig = plt.figure(figsize=(20, 10))

# Adding a 3D subplot to the figure
ax = fig.add_subplot(111, projection='3d')

# Creating a 3D scatter plot with age, annual income, and spending score as axes
ax.scatter(df.Age, df["Annual Income (k$)"], df["Spending Score (1-100)"], c='blue', s=60)

# Adjusting the viewing angle of the plot
ax.view_init(30, 185)

# Setting labels for x-axis, y-axis, and z-axis
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')

# Displaying the plot
plt.show()
```

```python
# Importing KMeans module from sklearn.cluster
from sklearn.cluster import KMeans

# Empty list to store the within-cluster sum of squares (WCSS) for different values of k
wcss = []

# Looping through values of k from 1 to 10
for k in range(1, 11):
    # Creating a KMeans model with k clusters and using k-means++ initialization
    kmeans = KMeans(n_clusters=k, init="k-means++")
    
    # Fitting the KMeans model to the data
    kmeans.fit(df.iloc[:,1:])
    
    # Appending the WCSS to the list
    wcss.append(kmeans.inertia_)

# Creating a new figure for the plot with a specific size
plt.figure(figsize=(12, 6))

# Creating a grid for better visualization
plt.grid()

# Plotting the WCSS values against the number of clusters (k)
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="8")

# Setting labels for x-axis and y-axis
plt.xlabel("K Value")
plt.ylabel("WCSS")

# Setting ticks for x-axis from 1 to 10
plt.xticks(np.arange(1, 11, 1))

# Displaying the plot
plt.show()
```

```python
# Importing necessary modules
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

# Creating a KMeans model with 5 clusters
km = KMeans(n_clusters=5)

# Fitting the KMeans model and predicting clusters
clusters = km.fit_predict(df.iloc[:,1:])

# Adding the cluster labels to the DataFrame
df["label"] = clusters

# Creating a new figure for the plot with a specific size
fig = plt.figure(figsize=(20, 10))

# Adding a 3D subplot to the figure
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each cluster
ax.scatter(df.Age[df.label == 0], df["Annual Income (k$)"][df.label == 0], df["Spending Score (1-100)"][df.label == 0], c='blue', s=60)
ax.scatter(df.Age[df.label == 1], df["Annual Income (k$)"][df.label == 1], df["Spending Score (1-100)"][df.label == 1], c='red', s=60)
ax.scatter(df.Age[df.label == 2], df["Annual Income (k$)"][df.label == 2], df["Spending Score (1-100)"][df.label == 2], c='green', s=60)
ax.scatter(df.Age[df.label == 3], df["Annual Income (k$)"][df.label == 3], df["Spending Score (1-100)"][df.label == 3], c='orange', s=60)
ax.scatter(df.Age[df.label == 4], df["Annual Income (k$)"][df.label == 4], df["Spending Score (1-100)"][df.label == 4], c='purple', s=60)

# Adjusting the viewing angle of the plot
ax.view_init(30, 185)

# Setting labels for x-axis, y-axis, and z-axis
ax.set_xlabel("Age")
ax.set_ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')

# Displaying the plot
plt.show()
```
